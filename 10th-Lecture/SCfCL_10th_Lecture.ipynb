## インポート
必要なライブラリやモジュールをインポートする

````python
import pandas as pd
````

## Seriesオブジェクト
- Series：一次元のコンテナの辞書型，単一の行や列としての一次元のデータ構造
- コンテナとは：オブジェクトの集まりを表現するデータ構造

````python
name_series = pd.Series(['田中', ' 北野'])
name_series
````

## DataFrameオブジェクト
- DataFrame
  - Excelのような表敬式のデータ型
  - 二次元データを表す
  - SeriesをまとめてDataFrameとして扱うこともできる

````python
people = pd.DataFrame({
    '名前' : name_series,
    '年齢' : [22, 28]})

people
````

## DataFrameから列単位で取り出す
DataFrameから列単位でSeriesとして取り出すことができる．キーを指定して取り出す

```python
people['年齢']
```
もしくは
```python
people.年齢
```
同じ結果になる

````python
people['年齢']
````

````python
people.年齢
````

## インデックスラベル（行名）
`Series`や`DataFrame`の引数に`index`を指定することで，デフォルトでない値を指定することができる

````python
name_series = pd.Series(['田中', '北野'], index = [10, 20])
name_series
````

## `loc`と`iloc`属性によるデータの選択
- loc属性
  - ラベルを用いて行や列にアクセスするための属性
  - 行や列のラベルを指定して特定のデータにアクセスする際に使用
- iloc属性
  - インデックスを用いて行や列にアクセスするための属性
  - 行や列の位置（インデックス）を指定して特定のデータにアクセスする際に使用

````python
# 事前準備
people = pd.DataFrame({
    '年齢' : [21, 34, 23, 44, 19],
    '身長' : [180, 158, 174, 181, 169],
    '体重' : [74, 61, 65, 82, 70]
},
    index = ['田中', '北野', '高橋', '岡田', '長谷川'])

people
````

`loc`属性でデータを選択と`iloc`属性でデータを選択

````python
# loc属性を利用する方法
people.loc['北野']
````

````python
# iloc属性を利用する方法
people.iloc[1]
````

必要な列に絞って取得することも可能

````python
# loc属性を利用する方法
people.loc['高橋', ['年齢', '体重']]
````

````python
# iloc属性を利用する方法
people.iloc[2, [0, 2]]
````

スライス記法`[start:end]`を使ってデータの取り出し

````python
# loc属性を利用する方法
people.loc['高橋':'長谷川']
````

````python
# iloc属性を利用する方法
people.iloc[2:5]
````
