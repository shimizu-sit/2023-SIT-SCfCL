```python
import openpyxl as op
from openpyxl.chart import BarChart, Reference, Series
# sin関数を計算するため
import math
```

## 問題1：世界の労働時間グラフ

Moodleから「workinghours.xlsx」をダウンロードしてColabにアップしてください．
以下の指示に従ってグラフを作成してください．
- 元データ：C3〜C44
- グラフの種類：棒グラフ
- グラフタイトル：労働時間
- 系列名：各国
- グラフ貼り付け位置セル：E3

実行結果例：https://github.com/shimizu-sit/2023-SIT-SCfCL/blob/main/7th-Lecture/SCfCL-7th-Prac-1-example.png?raw=true


```python
# workinghours.xlsxファイルを読み込みます
wb = op.load_workbook('/content/drive/MyDrive/SIT_2023/SCfCL/workinghours.xlsx')
# ワークシートを指定します
ws = wb.active

### データと系列の準備
# データの範囲を指定します
data = Reference(ws, min_col=3, min_row=3, max_col=3, max_row=44)
# 系列名を指定します
series = Series(data, title = '各国')

### グラフの準備
# 棒グラフを指定します
chart = BarChart()
# グラフタイトルを指定します
chart.title = '労働時間'
# グラフにデータ（系列）をわたします
chart.append(series)

### グラフの描画
# ワークシートのE3セルにグラフを作成します
ws.add_chart(chart, 'E3')

# エクセルファイルを保存します
wb.save('/content/drive/MyDrive/SIT_2023/SCfCL/SCfCL-7th-Prac-1.xlsx')
```

## 問題2：Excelファイルの作成とデータ準備

新規にExcelファイルを作成してください．
以下の指示に従ってExcelファイルにデータを作成してください．

- B2セル：「角度」，C3セル「SIN」
- B3セル〜B27セルに0〜360まで15刻みで数値入れる
- C3セル〜C27セルにB3セル〜B27セルの値を受け取りSINの値を入力

実行例：https://github.com/shimizu-sit/2023-SIT-SCfCL/blob/main/7th-Lecture/SCfCL-7th-Prac-2-example.png?raw=true

```python
# 新規にExcelファイルを作成します
wb = op.Workbook()
ws = wb.active

# 「角度」と「SIN]を入れます
ws['B2'].value = '角度'
ws['C2'].value = 'SIN'

# B3～B27に値を入れます
for i in range(0, 25):
  ws.cell(row = 3 + i, column=2).value = i * 15
  ws.cell(row = 3 + i, column=3).value = math.sin(math.radians(i * 15))


wb.save('/content/drive/MyDrive/SIT_2023/SCfCL/SCfCL-7th-Prac-2.xlsx')
```

## 問題3：棒グラフでSINグラフを作成

問題2で作成したExcelファイルを読み込んでください．以下の指示に従ってグラフを作成してください．

- 元データ：セルC3〜C27
- グラフの種類：棒グラフ
- グラフタイトル：三角関数
- 系列名：sin()
- グラフ貼り付け位置セル：E3

実行例：https://github.com/shimizu-sit/2023-SIT-SCfCL/blob/main/7th-Lecture/SCfCL-7th-Prac-3-example.png?raw=true


```python
### データと系列の準備
# データの範囲を指定します
data = Reference(ws, min_col=3, min_row=3, max_col=3, max_row=27)
# 系列名を指定します
series = Series(data, title = 'sin()')

### グラフの準備
# 棒グラフを指定します
chart = BarChart()
# グラフタイトルを指定します
chart.title = '三角関数'
# グラフにデータ（系列）をわたします
chart.append(series)

### グラフの描画
# ワークシートのE3セルにグラフを作成します
ws.add_chart(chart, 'E3')

# エクセルファイルを保存します
wb.save('/content/drive/MyDrive/SIT_2023/SCfCL/SCfCL-7th-Prac-3.xlsx')
```
