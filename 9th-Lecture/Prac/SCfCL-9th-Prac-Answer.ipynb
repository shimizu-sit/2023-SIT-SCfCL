## 問題5：CSVファイルの出力

「Prac09」フォルダに新しく「Prac09-3.csv」ファイルを作成してください．
「Prac09-3.csv」ファイルの中身は次のようにしてください．

**※アルファベットの順番に誤りがありますがそのままにしてください．**

```
A, B, C, D, E
F, G, H, I, J
K, L, N, M, O
P, Q, R, S, T
U, W, V, X, Y
Z,
```


````python
# フォルダとフォルダパスは設定されいているので書き込みから実行
with open(path + 'Prac09-3.csv', 'x', newline='') as f:
  w = csv.writer(f, delimiter=',')
  w.writerow(['A', 'B', 'C', 'D', 'E'])
  w.writerow(['F', 'G', 'H', 'I', 'J'])
  w.writerow(['K', 'L', 'N', 'M', 'O'])
  w.writerow(['P', 'Q', 'R', 'S', 'T'])
  w.writerow(['U', 'W', 'V', 'X', 'Y'])
  w.writerow(['Z'])
````

## 問題6：CSVファイルの加工

問題5で作成した「Prac09-3.csv」ファイルの中身には，アルファベットの順番に誤りがあります．
これを修正してください．


````python
#ファイルパスの設定
csv_file_path = path +'Prac09-3.csv'

# データの加工を行うのでpandasでファイル読み込み
df = pd.read_csv(csv_file_path, header=None)
df.loc[2] = ['K', 'L', 'M', 'N', 'O']
df.loc[4] = ['U', 'V', 'W', 'X', 'Y']
df.to_csv(path + 'Prac09-3-mod.csv', index=False, header=None)
````
